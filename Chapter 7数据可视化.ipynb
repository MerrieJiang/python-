## 7.3.2 金融图表

```python
import tushare as ts
```

```python
shlg = ts.get_hist_data('600848') #一次性获取上海临港全部日k线数据
```

```python
shlg.head()
```

```python
quotes = shlg[['open','high','low','close']]
quotes.head()
```

### 股票价格走势图

```python
qf = cf.QuantFig(quotes,title='600848 price information',legend = 'top',name='600848')#将dataframe对象作为输入
plyo.iplot(qf.iplot(asFigure=True),
          #Iimage='png',
           filename='qf_01'
          )
```

### 添加布林线

```python
qf.add_bollinger_bands(periods=15,boll_std=2)#设置布林带周期和标准差
plyo.iplot(qf.iplot(asFigure=True),
          #image='png',
          filename='qf_02'
          )
```

### 添加RSI指标（作为子图）

```python
qf.add_rsi(periods=14,showbands=False)#设置RSI周期，不显示高线或低线
plyo.iplot(qf.iplot(asFigure=True),
          #image='png',
          filename='qf_03')
```
